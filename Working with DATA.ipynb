```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
plt.scatter(df['Age'], df['Academic Pressure'],color="Blue",marker="*")
plt.title('Academic Pressure')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['Age'], df['Academic Pressure'],color="Blue",marker=".")
plt.xlabel('Age')
plt.ylabel('Academic Pressure')
# plt.plot(df['Age'],df['Academic Pressure'],color="Red")
# plt.show()
```

```python
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%0.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.show()
```

```python
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution - Count Plot')
plt.show()
```

```python
AcademicPressure_counts = df['Academic Pressure'].value_counts()
plt.pie(AcademicPressure_counts, labels=AcademicPressure_counts.index, autopct='%0.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Academic Pressure Distribution')
plt.show()
```

```python
sns.countplot(x='Financial Stress', data=df)
plt.title('Financial Stress Distribution - Count Plot')
plt.show()
```
